### Longest Palindrome Word
Input: <br>
madam racecar apple level python
<br>
Output: <br>
racecar

```python
x = "madam racecar apple level python".split()
max_length = 0
for i in x:
    if i == i[::-1]:
        if max_length < len(i):
            max_length = len(i)
            longest_word = i
            
print(longest_word)
```
